# Chicago Housing Data Analysis

```python
import pandas as pd 

df = pd.read_csv("Affordable_Rental_Housing_Developments.csv")
df
```

 # Count the number of developments per community area

```python
counts = df.groupby('Community Area Number')['Community Area Number'].count()
counts
```

 # Plot a histogram of the counts

```python
import matplotlib.pyplot as plt


plt.hist(counts, bins=20)
plt.xlabel('Number of Affordable Housing Developments')
plt.ylabel('Number of Community Areas')
plt.title('Distribution of Affordable Housing Developments in Chicago')
plt.show()
```

#Designing Boxplot for DataSet

```python
import seaborn as sns
sns.boxplot(x=counts)
plt.xlabel('Number of Affordable Housing Developments')
plt.title('Boxplot of Affordable Housing Developments in Chicago')
plt.show()
```

# Plot a bar chart of the counts by community area

```python
counts_by_area = df.groupby(['Community Area Name'])['Community Area Name'].count().sort_values()
counts_by_area.plot(kind='barh', figsize=(10,20), title='Number of Affordable Housing Developments by Community Area')
plt.xlabel('Number of Developments')
plt.show()
```
